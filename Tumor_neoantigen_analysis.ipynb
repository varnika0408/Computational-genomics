```python
#Analysis of tumour neoantigen analysis using MuPeXi for WD13


import pandas as pd

# Load MuPeXi results (Assuming it's a TSV file)
df = pd.read_csv("C:/Users/dell/Downloads/WD13_Vcf.txt", sep="\t", skiprows=5)


# Apply filters for tumor neoantigens
neoantigens = df[(df["Norm_MHCrank_EL"] > 2) & (df["Mut_MHCrank_EL"] < 2)]

# Apply filters for strong tumor neoantigens
strong_neoantigens = neoantigens[neoantigens["Mut_MHCrank_EL"] < 0.5]

# Count unique neoantigen peptides per tumor
neoantigen_counts = neoantigens.groupby("Gene_Symbol").size()

# Count strong neoantigens per tumor
strong_neoantigen_counts = strong_neoantigens.groupby("Gene_Symbol").size()

# Print results
print("Total Tumor Neoantigens per Tumor Sample of WD13:")
print(neoantigen_counts)

print("\nTotal Strong Tumor Neoantigens per Tumor Sample of WD13:")
print(strong_neoantigen_counts)

print(f"Total Tumour Neoantigens of WD13: {len(neoantigens)}")
print(f"Total Strong Tumour Neoantigens of WD13: {len(strong_neoantigens)}")

# Save results to Excel
neoantigens.to_excel("Neoantigens_WD13.xlsx", index=False)
strong_neoantigens.to_excel("Strong_Neoantigens_WD13.xlsx", index=False)
```

Neoantigen-based therapies utilize tumor-specific antigens derived from cancer mutations to stimulate immune responses. 

1). Positive Implications
1.1 Personalized Treatment:
Tailored to an individualâ€™s unique cancer mutations, enhancing efficacy while minimizing off-target effects.

1.2 Reduced Side Effects:
Since neoantigens are tumor-specific, they significantly lower the risk of autoimmune reactions and damage to healthy tissue.

1.3 Synergy with Other Therapies:
Can be combined with immunotherapies such as checkpoint inhibitors to enhance overall treatment effectiveness.

2). Challenges and Limitations
2.1 Complexity and Cost:
Requires extensive genomic analysis, bioinformatics assessment, and personalized manufacturing, increasing both cost and complexity.

2.2 Uncertain Immunogenicity:
Not all neoantigens elicit strong immune responses, making reliable identification a challenge.

2.3 Tumor Heterogeneity and Immune Evasion:
Variability within tumors may result in incomplete targeting, allowing some cancer cells to escape immune detection and potentially leading to recurrence.

2.4 Lengthy Development Process:
Personalized production and validation can take weeks or months, limiting accessibility for patients with rapidly progressing cancers.

2.5 Ethical and Accessibility Concerns:
High costs and the need for specialized facilities may restrict availability, exacerbating healthcare disparities.
